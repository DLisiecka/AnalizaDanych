```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

plt.style.use('seaborn-whitegrid')
plt.rcParams['figure.figsize'] = [10, 8]
plt.rcParams.update({'font.size': 15})
```

```python
df = pd.read_excel('DATA/1.Wypadki_ogolem_smiertleni_ranni_WOJ.xlsx')
```

```python
### Statystyka wypadków, ofiar śmiertelnych, rannych w Polsce w latach 2014-2020
```

```python
df_pl = df[ df['Jednostka terytorialna'] == 'POLSKA']
```

```python
df_pl[['Wypadki ogółem','Ofiary śmiertelne','Ranni']].aggregate(['sum','mean'])
```

```python
df_pl.sort_values('Wypadki ogółem',ascending=False)
```

```python
df_pl.sort_values('Ofiary śmiertelne',ascending=False)
```

```python
df_pl.sort_values('Ranni',ascending=False)
```

```python
#Wykres ogólnej liczby wypadków w Polsce w latach 2014-2020

sns.scatterplot(data = df_pl, x='Rok', y='Wypadki ogółem')
plt.show()
```

```python
### Statystyka wypadków, ofiar śmiertelnych, rannych w poszczególnych województwach w Polsce w latach 2014-2020
```

```python
## Wypadki ogółem 

df_woj = df[ df['Jednostka terytorialna'] != 'POLSKA']
df_wypadki = df_woj.groupby('Jednostka terytorialna')['Wypadki ogółem'].agg(['max','min','sum','mean'])
```

```python
df_wypadki_sum = df_wypadki.sort_values('sum', ascending=False)
df_wypadki_sum
```

```python
df_wypadki_max = df_wypadki.sort_values('max', ascending=False)
df_wypadki_max
```

```python
df_wypadki_min = df_wypadki.sort_values('min')
df_wypadki_min
```

```python
df_wide = df_woj.pivot("Rok", "Jednostka terytorialna", "Wypadki ogółem")
df_wide
```

```python
sns.lineplot(data=df_wide).set(title='Liczba wypadków ogółem w latach 2014-2020')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0, title='Jednostka terytorialna')
plt.show()
```

```python
## Ofiary śmiertelne
```

```python
df_ofiary = df_woj.groupby('Jednostka terytorialna')['Ofiary śmiertelne'].agg(['max','min','sum','mean'])
```

```python
df_ofiary_sum = df_ofiary.sort_values('sum', ascending=False)
df_ofiary_sum
```

```python
df_ofiary_max = df_ofiary.sort_values('max', ascending=False)
df_ofiary_max
```

```python
df_ofiary_min = df_ofiary.sort_values('min')
df_ofiary_min
```

```python
df_wide_2 = df_woj.pivot("Rok", "Jednostka terytorialna", "Ofiary śmiertelne")
df_wide_2
```

```python
sns.lineplot(data=df_wide_2).set(title='Liczba ofiar śmiertelnych w latach 2014-2020')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0, title='Jednostka terytorialna')
plt.show()
```

```python
## Ranni
```

```python
df_ranni = df_woj.groupby('Jednostka terytorialna')['Ranni'].agg(['max','min','sum','mean'])
```

```python
df_ranni_sum = df_ranni.sort_values('sum', ascending=False)
df_ranni_sum
```

```python
df_ranni_max = df_ranni.sort_values('max', ascending=False)
df_ranni_max
```

```python
df_ranni_min = df_ranni.sort_values('min')
df_ranni_min
```

```python
df_wide_3 = df_woj.pivot("Rok", "Jednostka terytorialna", "Ranni")
df_wide_3
```

```python
sns.lineplot(data=df_wide_3).set(title='Liczba rannych w latach 2014-2020')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0, title='Jednostka terytorialna')
plt.show()
```
